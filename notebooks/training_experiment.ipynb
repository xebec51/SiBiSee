```python
# --- SEL 2: TRAINING BASELINE ---

from ultralytics import YOLO

# 1. Tentukan path file konfigurasi data.yaml
# Path yang benar berdasarkan direktori kerja (ultralytics) dan folder datamu (SIBI-2)
DATA_YAML_PATH = "./SIBI-2/data.yaml"

# Cek keberadaan file sebelum training (Verifikasi akhir)
import os
if not os.path.exists(DATA_YAML_PATH):
    raise FileNotFoundError(f"File data.yaml tidak ditemukan pada path: {DATA_YAML_PATH}")
else:
    print(f"✅ File data.yaml Ditemukan di: {DATA_YAML_PATH}")

# 2. Muat Model YOLOv8s (small) Pre-trained dari COCO
# File yolov8s.pt sudah didownload
model = YOLO("yolov8s.pt")

# 3. Training Model Baseline
print("\n--- MEMULAI TRAINING BASELINE: YOLOv8s ---")
# Setting: epochs=25, batch=16, device=0 (RTX 4060 Ti)
results_baseline = model.train(
    data=DATA_YAML_PATH, 
    epochs=25,
    imgsz=640,
    batch=16, 
    name='yolov8s_baseline',
    device=0
)

# Tampilkan hasil metrik utama
print("\n--- HASIL TRAINING BASELINE SELESAI ---")
print(results_baseline.metrics)
```

```python
# --- SEL 3: AUTO-CREATE CONFIG & TRAIN (FINAL FIX SCALING) ---
import os
from ultralytics import YOLO

# ==========================================
# BAGIAN 1: MEMBUAT FILE KONFIGURASI (YAML)
# ==========================================

yaml_content = """
# Ultralytics YOLO, AGPL-3.0 license
# YOLOv8s-CBAM Improved Model Architecture

# Parameters
nc: 26  # SIBI (A-Z)
scales:
  n: [0.33, 0.25, 1024]
  s: [0.33, 0.50, 1024] # Note: Width scale is 0.50
  m: [0.67, 0.75, 768]
  l: [1.00, 1.00, 512]
  x: [1.00, 1.25, 512]

backbone:
  # [from, repeats, module, args]
  - [-1, 1, Conv, [64, 3, 2]]  # 0-P1/2
  - [-1, 1, Conv, [128, 3, 2]]  # 1-P2/4
  - [-1, 3, C2f, [128, True]]
  - [-1, 1, Conv, [256, 3, 2]]  # 3-P3/8
  - [-1, 6, C2f, [256, True]]
  - [-1, 1, Conv, [512, 3, 2]]  # 5-P4/16
  - [-1, 6, C2f, [512, True]]  # 6 (Output asli: 512 * 0.5 = 256 channels)
  
  # >>> MODIFIKASI KITA: CBAM
  # KOREKSI: Gunakan 256, bukan 512. 
  # Karena layer sebelumnya (C2f) kena scaling 0.5, outputnya jadi 256.
  - [-1, 1, CBAM, [256]]  # 7 <-- CBAM disesuaikan dengan input 256
  
  - [-1, 1, Conv, [1024, 3, 2]]  # 8-P5/32
  - [-1, 3, C2f, [1024, True]]  # 9
  - [-1, 1, SPPF, [1024, 5]]  # 10

head:
  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]
  - [[-1, 6], 1, Concat, [1]]  # cat backbone P4
  - [-1, 3, C2f, [512]]  # 12

  - [-1, 1, nn.Upsample, [None, 2, "nearest"]]
  - [[-1, 4], 1, Concat, [1]]  # cat backbone P3
  - [-1, 3, C2f, [256]]  # 15 (P3/8-small)

  - [15, 1, Conv, [256, 3, 2]]
  - [[-1, 12], 1, Concat, [1]]  # cat head P4
  - [-1, 3, C2f, [512]]  # 18 (P4/16-medium)

  - [-1, 1, Conv, [512, 3, 2]]
  - [[-1, 9], 1, Concat, [1]]  # cat head P5
  - [-1, 3, C2f, [1024]]  # 21 (P5/32-large)

  - [[15, 18, 21], 1, Detect, [nc]]  # Detect(P3, P4, P5)
"""

custom_yaml_path = os.path.join(os.getcwd(), "yolov8s-cbam-custom.yaml")

# Simpan dengan encoding utf-8
with open(custom_yaml_path, "w", encoding="utf-8") as f:
    f.write(yaml_content)

print(f"✅ File konfigurasi BERHASIL dibuat di: {custom_yaml_path}")


# ==========================================
# BAGIAN 2: TRAINING MODEL
# ==========================================

print("⏳ Memuat model Improved...")
model_improved = YOLO(custom_yaml_path)

DATA_YAML_PATH = "./SIBI-2/data.yaml" 

print("\n--- MEMULAI TRAINING MODEL IMPROVED: YOLOv8s-CBAM ---")
try:
    # Kita coba transfer learning, tapi jika layer shape beda (karena ada CBAM), 
    # YOLO akan otomatis skip bobot yang tidak cocok dan load sisanya. Aman.
    results_improved = model_improved.train(
        data=DATA_YAML_PATH,
        epochs=25,
        imgsz=640,
        batch=16, 
        name='yolov8s_cbam_final',
        device=0,
        weights='runs/detect/yolov8s_baseline2/weights/best.pt' 
    )
except Exception as e:
    print(f"⚠️ Warning Transfer Learning: {e}")
    print("🔄 Beralih ke training dari awal (scratch)...")
    results_improved = model_improved.train(
        data=DATA_YAML_PATH,
        epochs=25,
        imgsz=640,
        batch=16, 
        name='yolov8s_cbam_final_scratch',
        device=0
    )

print("\n--- HASIL TRAINING IMPROVED SELESAI ---")
print(results_improved.results_dict)
```
